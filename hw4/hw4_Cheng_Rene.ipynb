```python
import matplotlib.pyplot as plt
import numpy as np
import pylab

x1,x2 = np.meshgrid(np.linspace(-10,20,60),np.linspace(-10,20,60))
x= np.linspace(-10,10,1000)
y= np.linspace(-10,10,1000)
dp = ((10*(x1**2))/(4+(x1**2))-(2*x1))
plt.figure()

q = plt.quiver(x1,x2, dp, 0, scale=35, width=0.003)
plt.plot(x,((10*(x**2))/(4+(x**2))-(2*x)))
pylab.title("pmax > 2kp*rp")

pylab.xlabel("[P]")
pylab.xlabel("dp/dt")
pylab.ylim([-3,4])
pylab.xlim([-3,8])
pylab.axhline(y=0)
plt.show()
print ("Pmax=10, k=2, rp=2")
```
