```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
from sympy.solvers import solve
from sympy import Symbol

A, R = np.meshgrid(np.arange(0,250,8), np.arange(0,250,8))

alphA = 250
alphR = 30

alphA0 = 0.01
alphR0 = 0
gammA = 1
gammR = 0.5


n = 1
m = 2

dA = ((alphA*A**n+alphA0)/(1+A**n+R**m))-gammA*A
dR = ((alphR*A**n+alphR0)/(1+A**n))-gammR*R

#x-nullcline
x = np.arange(0,250,1)
xnull = (((alphA*x**n+alphA0)/(gammA*x))-1-x**n)**(1/m)
plt.plot(x , xnull, label="x-nullcline")

#y-nullcline
y = np.arange(0,250,1)
ynull = (alphR*y**n+alphR0)/(gammR*(1+y**n))
plt.plot(y, ynull, label="y-nullcline")

#normalize
dAnorm = dA/(dA**2+dR**2)**.5 *300
dRnorm = dR/(dA**2+dR**2)**.5 *300

plt.xlim(0,250)
plt.ylim(0,100)
plt.quiver(A, R, dAnorm, dRnorm,scale = 10000, width = 0.003)
plt.xlabel('[A] in nM')
plt.ylabel('[R] in nM')
plt.legend()
plt.show()
```

```python
alphA = 250
alphR = 30

alphA0 = 0.01
alphR0 = 0
gammA = 1
gammR = 0.5

n = 1
m = 2

t0 = 0
tn = 30
tsteps = 1001
t = np.linspace(t0, tn, tsteps)

def get_ode_derivatives(c, t):
    A, R = c
    
    dcdt = [
        ((alphA*(A**n)+alphA0)/(1+(A**n)+(R**m)))-gammA*A,
        ((alphR*(A**n)+alphR0)/(1+(A**n)))-gammR*R
    ]
    
    return dcdt

sol = odeint(get_ode_derivatives, (1,1), t)
plt.plot (t, sol[:, 0], 'b', label = "A")
plt.plot (t, sol[:, 1], 'g', label = "R")
plt.legend()
plt.xlabel('t')
plt.show()
```
